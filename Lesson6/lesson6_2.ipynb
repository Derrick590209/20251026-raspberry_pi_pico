# MQTT Subscriber 測試程式

這個 notebook 示範如何使用 `paho-mqtt` 建立一個簡單的 MQTT 訂閱者（Subscriber）應用程式，用來接收並顯示從 MQTT Broker 收到的訊息。


```python
import time
import json
import random

import paho.mqtt.client as mqtt

# MQTT Broker 設定（改成連到本機 Raspberry Pi 上的 Broker）
BROKER_HOST = "localhost"  # 或 "127.0.0.1"，如果在同一台機器上
BROKER_PORT = 1883          # 若你有改 Mosquitto 的埠號，就在這裡同步修改
TOPIC = "客廳/溫度"    # 要訂閱的主題（需與 Publisher 使用相同的主題）
CLIENT_ID = f"subscriber_test_{random.randint(1000, 9999)}"  # 隨機客戶端 ID 避免衝突

print("MQTT Subscriber 測試設定完成")
print(f"Broker: {BROKER_HOST}:{BROKER_PORT}")
print(f"Topic: {TOPIC}")
print(f"Client ID: {CLIENT_ID}")
```

```python
# 建立 MQTT 客戶端並連線到 Broker

client = mqtt.Client(client_id=CLIENT_ID)

connected = False
messages_received = []  # 用來儲存收到的訊息

# 連線回呼函數
def on_connect(client, userdata, flags, rc):
    global connected
    if rc == 0:
        connected = True
        print("✓ 成功連接到 MQTT Broker")
        print(f"  - 連線標誌: {flags}")
    else:
        connected = False
        error_messages = {
            1: "協議版本不正確",
            2: "客戶端 ID 無效",
            3: "伺服器不可用",
            4: "用戶名或密碼錯誤",
            5: "未授權",
        }
        error_msg = error_messages.get(rc, "未知錯誤")
        print(f"✗ 連接失敗，錯誤代碼 {rc}: {error_msg}")


# 訂閱成功回呼函數
def on_subscribe(client, userdata, mid, granted_qos):
    print(f"✓ 成功訂閱主題 (Message ID: {mid}, QoS: {granted_qos})")


# 接收訊息回呼函數
def on_message(client, userdata, msg):
    topic = msg.topic
    payload = msg.payload.decode('utf-8')
    timestamp = time.strftime('%Y-%m-%d %H:%M:%S')
    
    print(f"\n[{timestamp}] 收到訊息:")
    print(f"  主題: {topic}")
    print(f"  內容: {payload}")
    
    # 嘗試解析 JSON 格式
    try:
        json_data = json.loads(payload)
        print(f"  JSON 解析成功:")
        for key, value in json_data.items():
            print(f"    {key}: {value}")
    except json.JSONDecodeError:
        print(f"  (純文字訊息，非 JSON 格式)")
    
    # 儲存訊息
    messages_received.append({
        "timestamp": timestamp,
        "topic": topic,
        "payload": payload
    })


client.on_connect = on_connect
client.on_subscribe = on_subscribe
client.on_message = on_message

print(f"\n正在連接到 {BROKER_HOST}:{BROKER_PORT}...")
try:
    client.connect(BROKER_HOST, BROKER_PORT, keepalive=60)
    client.loop_start()  # 開始背景處理
    time.sleep(2)        # 留一點時間讓連線完成

    if connected:
        # 訂閱主題
        result, mid = client.subscribe(TOPIC, qos=1)
        if result == mqtt.MQTT_ERR_SUCCESS:
            print(f"✓ 正在訂閱主題: {TOPIC}")
            time.sleep(1)  # 等待訂閱確認
        else:
            print(f"✗ 訂閱失敗，錯誤代碼: {result}")
        print("✓ 準備就緒，開始監聽訊息...")
        print("  (請在另一個 notebook 或應用程式中發布訊息到此主題)")
    else:
        print("⚠ 連線狀態未知，請檢查網路或 Broker 設定")
except Exception as e:
    print(f"✗ 連接錯誤: {e}")
    print("  請檢查：")
    print("  1. 網路連線是否正常")
    print("  2. Broker 地址是否正確")
    print("  3. 防火牆是否阻擋了連接")
```

```python
# 持續監聽訊息（執行此 cell 後，會持續接收訊息）
# 要停止時，請中斷執行或執行下一個 cell

print("開始監聽訊息...")
print("(可以執行 30 秒來測試，或隨時中斷執行)")

try:
    # 持續監聽 30 秒（你可以改成更長時間或改成無限循環）
    time.sleep(30)
    print("\n✓ 監聽時間結束")
except KeyboardInterrupt:
    print("\n✓ 已停止監聽")
```

```python
# 查看已收到的訊息總數

print(f"總共收到 {len(messages_received)} 條訊息")
print("\n收到的訊息列表:")
print("-" * 50)

for i, msg in enumerate(messages_received, 1):
    print(f"\n訊息 #{i}:")
    print(f"  時間: {msg['timestamp']}")
    print(f"  主題: {msg['topic']}")
    print(f"  內容: {msg['payload']}")
```

```python
# 關閉 MQTT 連線

client.loop_stop()  # 停止背景處理
client.disconnect()
print("✓ MQTT 連線已關閉")
```

## 使用方式說明

1. **依序執行上面的 cells**：
   - Cell 1：匯入套件並設定 Broker / Topic / Client ID
   - Cell 2：建立 MQTT 客戶端、連線並訂閱主題
   - Cell 3：開始監聽訊息（預設監聽 30 秒，可以中斷執行）
   - Cell 4：查看已收到的訊息列表
   - Cell 5：關閉連線

2. **測試方式**：
   - 保持 Cell 2 和 Cell 3 執行後，在另一個終端機或 notebook（例如 `lesson6_1.ipynb`）發布訊息到相同的主題
   - 或使用命令行工具：`mosquitto_pub -h localhost -t "客廳/溫度" -m "測試訊息"`

3. **主題設定**：
   - 確保 `TOPIC` 與 Publisher 使用的主題相同（目前設定為 `"客廳/溫度"`）
   - 若想訂閱多個主題，可以使用萬用字元：
     - `"home/+/temperature"` - 訂閱所有符合此模式的主題
     - `"home/#"` - 訂閱所有以 `home/` 開頭的主題

4. **持續監聽**：
   - 若要持續監聽（無限時間），可以在 Cell 3 中將 `time.sleep(30)` 改成 `while True: time.sleep(1)`
   - 記得要手動中斷執行才能停止
